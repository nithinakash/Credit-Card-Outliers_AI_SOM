### Packages Required

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
```

### Datasets

```python
path = 'C:/Users/Nithin/Downloads/Credit Card Outliers_AI_SOM/'
dataset = pd.read_csv(path + 'Datasets/Credit_Card_Applications.csv')
dataset.headad(5)
```

```python
dataset.info()
```

```python
X = dataset.iloc[:, 1:14].values
y = dataset.iloc[:, -1].values
```

```python
pd.DataFrame(X)
```

```python
sc = MinMaxScaler(feature_range = (0, 1))
X = sc.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
#pip install minisom
from minisom import MiniSom
```

```python
som = MiniSom(x = 10, y = 10, input_len = 13, sigma = 1.0, learning_rate = 0.5)
som.random_weights_init(X)
som.train_random(data = X, num_iteration = 20000)
```

```python
som.get_weights()
```

```python
som.distance_map()
```

```python
from pylab import bone, pcolor, colorbar, plot, show
```

```python
pcolor(som.distance_map().T)
colorbar()
```

```python
pcolor(som.distance_map().T)
colorbar() #gives legend

markers = ['o', 's']
colors = ['r', 'g']
for i, x in enumerate(X):
    w = som.winner(x)
    plot(w[0] + 0.5,
         w[1] + 0.5,
         markers[y[i]],
         markeredgecolor = colors[y[i]],
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2)
show()
```

#### Tuning parameters
    -x = 10, 
    -y = 10, 
    -sigma = 1.0, 
    -learning_rate = 0.5
